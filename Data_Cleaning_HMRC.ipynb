***Reasons for Including Each Dataset***

**Hypothesis:**
I believe there may be a correlation between the tax gap and rising income over time in the UK.

**Dataset 1** — HMRC Measuring Tax Gaps (Primary Dataset)
Source:  HM Revenue & Customs — GOV.UK
File: Measuring_tax_gap_online_tables_2025.xlsx
URL: https://www.gov.uk/government/statistics/measuring-tax-gaps-tables

**Reason for inclusion:**
This is the primary metric under investigation. The tax gap measures the difference between the tax theoretically owed and the amount actually collected. It is HMRC’s own headline compliance indicator, published annually since 2005–06. Over the 19-year series it has declined from 7.4% of theoretical liabilities in 2005–06 to 5.3% in 2023–24, making it a clear and well-sourced ‘changing value’ for our purposes.


**Dataset 2** — ONS Annual Survey of Hours and Earnings (ASHE)
Source:  Office for National Statistics — ONS Website
File: selectedestimates19972025.xlsx
URL: https://www.ons.gov.uk/employmentandlabourmarket/peopleinwork/earningsandworkinghours/datasets/ashe1997to2015selectedestimates

**Reason for inclusion:**
ASHE provides the UK’s most comprehensive and consistent annual series of employee earnings, running from 1997 to 2025.

***Dashboard Insights***

**What the axes show** 
Each dot is one year. Its horizontal position (x-axis) is that year's median annual earnings, and its vertical position (y-axis) is the total tax gap in £bn. So dots further right = higher wages, dots further up = bigger tax gap.

**The overall trend** The dots form a rough upward slope from left to right — as earnings have grown over time, the tax gap has also generally grown. 2023 and 2024 sit at the top right, with both the highest wages and the biggest tax gap (around £46-47bn).

**The colour (Income YoY %)** 
The colour scale shows how fast income grew that year. Yellow/orange dots (2022, 2023, 2024) had fast income growth — those are the post-COVID years where wages surged. Dark purple dots (2010, 2012) had near-zero or negative income growth — the austerity years after the 2008 financial crisis.

**Outliers:**
*2020 sits unusually high for its earnings level — the tax gap spiked during COVID despite wages barely moving, likely due to disrupted tax collection
*2010 is the lowest dot overall — earnings actually fell slightly that year (dark purple) and the tax gap was at its smallest
*2014 appears above the trend line — the tax gap jumped to £36.5bn while earnings were still relatively modest at £22k

**Conclusion** 
There is a loose correlation that requires further research. Earnings and the tax gap have broadly risen together, but the tax gap has accelerated sharply since 2022 while income growth, though faster, hasn't kept pace.

Importing the two datasets

```python
# Adjust the path if your username is different
income_raw = pd.read_excel('/Users/alexandruvasile/Desktop/selectedestimates19972025.xlsx',
    sheet_name="Table 1",   # All employees pay
    header=None             # Headers to be handled manually
)

# Preview the raw data to see its structure
income_raw.head(15)
```

```python
taxgap_raw = pd.read_excel(
    '/Users/alexandruvasile/Desktop/Measuring_tax_gap_online_tables_2025.xlsx',
    sheet_name=  "Table 1.3",   # Tax gap in £ billion by type of tax
    header=None
)

# Preview
taxgap_raw.head(10)
```

--- INCOME DATA CLEANING ---

Step 1: Pull out the year row (row index 2) and earnings row (row index 6)

```python
year_row      = income_raw.iloc[2]   # e.g. 1997, 1998, ... 2025
earnings_row  = income_raw.iloc[6]   # median gross annual earnings values
```

Step 2: Build a small dataframe from those two rows

```python
income_df = pd.DataFrame({
    'year':     year_row.values,
    'median_annual_earnings_gbp': earnings_row.values
})
```

Step 3: Drop rows where the year is not a plain number

```python
# (removes NaNs, "2004 exc.", "2021Soc20" duplicates etc.)
income_df = income_df[pd.to_numeric(income_df['year'], errors='coerce').notna()]
```

Step 4: Convert year to integer and earnings to float

```python
income_df['year'] = income_df['year'].astype(int)
income_df['median_annual_earnings_gbp'] = pd.to_numeric(
    income_df['median_annual_earnings_gbp'], errors='coerce'
)
```

Step 5: Drop any remaining rows with no earnings value

```python
income_df = income_df.dropna(subset=['median_annual_earnings_gbp'])
```

Step 6: If any year appears twice (due to methodology breaks), keep the last one

```python
income_df = income_df.drop_duplicates(subset='year', keep='last')

income_df = income_df.reset_index(drop=True)
print(income_df)
```

--- TAX GAP DATA CLEANING ---

Step 1: Row 4 is the real header — use it, then re-read from row 5 onwards

```python
taxgap_df = taxgap_raw.copy()
taxgap_df.columns = taxgap_raw.iloc[4]     # set row 4 as column names
taxgap_df = taxgap_df.iloc[5:].reset_index(drop=True)  # drop the header rows
```

Step 2: Find the "Total tax gap" row

```python
total_row = taxgap_df[taxgap_df['Tax'] == 'Total tax gap']
```

Step 3: Keep only the year columns (drop 'Tax' and 'Type')

```python
total_row = total_row.drop(columns=['Tax', 'Type'])
```

Step 4: Transpose so each year becomes a row

```python
taxgap_clean = total_row.T.reset_index()
taxgap_clean.columns = ['tax_year', 'total_tax_gap_bn']
```

Step 5: Convert "2005-06" → 2006 (use the end year to match income data)

```python
taxgap_clean['year'] = taxgap_clean['tax_year'].str.split('-').str[1].astype(int) + 2000
```

Step 6: Convert tax gap to float (the "<0.1" entries won't appear in the total row)

```python
taxgap_clean['total_tax_gap_bn'] = pd.to_numeric(
    taxgap_clean['total_tax_gap_bn'], errors='coerce'
)

taxgap_clean = taxgap_clean[['year', 'total_tax_gap_bn']].reset_index(drop=True)
print(taxgap_clean)
```

--- MERGE INTO ONE DATAFRAME ---

```python
# Inner join on year — keeps only the overlapping years (2006–2024)
dashboard_df = pd.merge(income_df, taxgap_clean, on='year', how='inner')

# Calculate year-on-year % changes for both metrics
dashboard_df['income_yoy_pct'] = dashboard_df['median_annual_earnings_gbp'].pct_change() * 100
dashboard_df['taxgap_yoy_pct'] = dashboard_df['total_tax_gap_bn'].pct_change() * 100

dashboard_df = dashboard_df.round(2)
print(dashboard_df)
print(f"\nShape: {dashboard_df.shape}")
print(f"Years covered: {dashboard_df['year'].min()} – {dashboard_df['year'].max()}")

# Export to Desktop
dashboard_df.to_csv('/Users/alexandruvasile/Desktop/dashboard_data.csv', index=False)
print("CSV saved to Desktop!")
```
